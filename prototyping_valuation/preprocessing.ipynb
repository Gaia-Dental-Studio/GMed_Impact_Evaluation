```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('prototyping_cost.csv')
data
```

```python
# make column 'Cost ID' to be the index and set the value to be from 1 until the end of the dataframe
data.set_index('Cost ID', inplace=True)
data.index = np.arange(1, len(data) + 1)
data
```

```python
# For 'Category' column, for every row that has no value, fill it with the value from the previous row
data['Category'].fillna(method='ffill', inplace=True)
```

```python
# for column 'Amount Low', 'Amount High', 'Amount AUD Low', and 'Amount AUD High', delete any strings that are not numeric, including commas
def clean_numeric_column(column):
    column = column.str.replace(r'[^\d.]', '', regex=True)
    return pd.to_numeric(column, errors='coerce')

data['Amount Low'] = clean_numeric_column(data['Amount Low'])
data['Amount High'] = clean_numeric_column(data['Amount High'])
data['Amount AUD Low'] = clean_numeric_column(data['Amount AUD Low'])
data['Amount AUD High'] = clean_numeric_column(data['Amount AUD High'])

data
```

```python
# drop rows where 'Sub-Category' is 'Hardware'
data = data[data['Relative to Prototype Number'] != 'Variable Cost']
```

```python
data.to_csv('prototyping_cost_cleaned.csv', index=True)
```

```python
# groupby 'Category' and calculate the sum for 'Amount Low', 'Amount High', 'Amount AUD Low', and 'Amount AUD High'
grouped_data = data.groupby('Category').agg({
    'Amount Low': 'sum',
    'Amount High': 'sum',
    'Amount AUD Low': 'sum',
    'Amount AUD High': 'sum'
}).reset_index()
```
