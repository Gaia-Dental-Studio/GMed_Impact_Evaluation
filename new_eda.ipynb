```python
import pandas as pd
import numpy as np

data = pd.read_csv('world_data_new.csv')
data
```

```python
# erase last 7 column
data = data.iloc[:, :-7]

data
```

```python
# locate sixth column, disregard the '%' and convert to float divided by 100
data.iloc[:, 5] = data.iloc[:, 5].str.rstrip('%').astype(float) / 100
data.iloc[:, 8] = data.iloc[:, 8].str.rstrip('%').astype(float) / 100
data.iloc[:, -3] = data.iloc[:, -3].str.rstrip('%').astype(float) / 100
data.iloc[:, -1] = data.iloc[:, -1].str.rstrip('%').astype(float) / 100
data
```

```python
import re

def parse_billion(val):
    if pd.isnull(val):
        return np.nan
    # Remove commas and spaces
    val = str(val).replace(',', '').strip()
    # Extract numeric part
    match = re.match(r"([0-9.]+)", val)
    if not match:
        return np.nan
    num = float(match.group(1))
    # Check if 'B' or 'b' is present (for billion)
    if 'b' in val.lower():
        num *= 1_000_000_000
    return num

col_idx = -2

data.iloc[:, col_idx] = data.iloc[:, col_idx].apply(parse_billion)
data
```

```python
col_idx = 7

data.iloc[:, col_idx] = data.iloc[:, col_idx].apply(parse_billion)
data
```

```python
# Delete rows where both the last and second last columns are NaN or 0
last_col = data.columns[-1]
second_last_col = data.columns[-2]

mask = ~(
    (data[last_col].isna() | (data[last_col] == 0)) &
    (data[second_last_col].isna() | (data[second_last_col] == 0))
)

data = data[mask]
data
```

```python
last_col = data.columns[-3]
second_last_col = data.columns[-4]

mask = ~(
    (data[last_col].isna() | (data[last_col] == 0)) &
    (data[second_last_col].isna() | (data[second_last_col] == 0))
)

data = data[mask]
data
```

```python
last_col = data.columns[8]
second_last_col = data.columns[9]

mask = ~(
    (data[last_col].isna() | (data[last_col] == 0)) &
    (data[second_last_col].isna() | (data[second_last_col] == 0))
)

data = data[mask]
data
```

```python
# delete the commas from column 3,4 and 6 and convert to float
data.iloc[:, 3] = data.iloc[:, 3].str.replace(',', '').astype(float)
data.iloc[:, 4] = data.iloc[:, 4].str.replace(',', '').astype(float)
data.iloc[:, 6] = data.iloc[:, 6].str.replace(',', '').astype(float)
```

```python
first_col = data.columns[5]
second_col = data.columns[6]
third_col = data.columns[4]

mask = data[first_col].isna() | (data[first_col] == 0) | (data[first_col] == '') 
data.loc[mask, first_col] = data.loc[mask, second_col] / data.loc[mask, third_col]

data
```

```python
first_col = data.columns[8]
second_col = data.columns[9]
third_col = data.columns[6]

mask = data[first_col].isna() | (data[first_col] == 0) | (data[first_col] == '') 
data.loc[mask, first_col] = data.loc[mask, second_col] * data.loc[mask, third_col]

data
```

```python
first_col = data.columns[-4]
second_col = data.columns[-3]
third_col = data.columns[6]

mask = data[first_col].isna() | (data[first_col] == 0) | (data[first_col] == '') 
data.loc[mask, first_col] = data.loc[mask, second_col] * data.loc[mask, third_col]

data
```

```python
first_col = data.columns[-2]
second_col = data.columns[-1]
third_col = data.columns[7]

mask = data[first_col].isna() | (data[first_col] == 0) | (data[first_col] == '') 
data.loc[mask, first_col] = data.loc[mask, second_col] * data.loc[mask, third_col]

data
```

```python
first_col = data.columns[-3]
second_col = data.columns[-4]
third_col = data.columns[4]

mask = data[first_col].isna() | (data[first_col] == 0) | (data[first_col] == '') 
data.loc[mask, first_col] = data.loc[mask, second_col] / data.loc[mask, third_col]

data
```

```python
first_col = data.columns[-3]
second_col = data.columns[-4]
third_col = data.columns[6]

mask = data[first_col].isna() | (data[first_col] == 0) | (data[first_col] == '') 
data.loc[mask, first_col] = data.loc[mask, second_col] * data.loc[mask, third_col]

data
```

```python
# Keep only the row with the highest value in column 2 for each unique combination of columns 0 and 1
data_unique = data.sort_values(by=data.columns[2], ascending=False).drop_duplicates(subset=[data.columns[0], data.columns[1]], keep='first')
data_unique
```

```python
# drop if first column is 'Indonesia' 
data_unique = data_unique[data_unique.iloc[:, 0] != 'Indonesia']
data_unique
```

```python
# rename 'Singapure' to 'Singapore' in 'Country' column
data_unique['Country'] = data_unique['Country'].replace('Singapure', 'Singapore')
```

```python
data_unique.iloc[:, 7] = data_unique.iloc[:, 7] * data_unique.iloc[:, 4] / data_unique.iloc[:, 3]
```

```python
data_unique.to_csv('world_data_cleaned_new.csv', index=False)
```
